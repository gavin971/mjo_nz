# Some comparison of filtering procedures for the Southern Annular Mode index

```python
%matplotlib inline
```

```python
import numpy as np
from matplotlib import pyplot as plt
import scipy.signal as signal
from scipy.io.matlab import loadmat
import scipy.ndimage as nd
import pandas as pd
```

### Get the raw SAM index from the Climate Prediction Center

```python
def get_SAM(url='ftp://ftp.cpc.ncep.noaa.gov/cwlinks/norm.daily.aao.index.b790101.current.ascii',start=1979, end=2013):
    data = pd.read_table(url, header=None, sep="\s*", 
                         names=['year','month','day','sam'], engine='python')
    data.index = pd.date_range(start="1979-1-1", periods=len(data))
    data = data.truncate(before="%s-1-1" % ( start ), after= "%s-12-31" % ( end ))
    data.fillna(data.mean(), inplace=True)
    return data
```

```python
SAM = get_SAM()
```

```python
SAM.head()
```

### Loads the SAM filtered via Matlab's butterworth function

```python
sam_filtered = loadmat('../../data/sam_filtered_20_120_filtbut.mat', struct_as_record=False)
sam_filtered = sam_filtered['sam_filtered']
```

### Filter the raw SAM as downloaded from the CPC using a Gaussian filter (sigma=3)

```python
sigma = 3
sam_gauss = nd.gaussian_filter1d(SAM['sam'].values, sigma)
```

### Filter the SAM using scipy.signal butter and filtfilt functions

```python
high = 20 # first cut-off period given in time_unit
low = 120  
```

```python
N  = 2.      # Filter order
```

```python
B, A = signal.butter(N, [N/low, N/high], btype='bandpass', output='ba')
```

```python
sam_filt_butter = signal.filtfilt(B, A, SAM['sam'].values)
```

### Plots the comparison

```python
f, ax = plt.subplots(figsize=(10,8))
ax.plot(SAM['sam'].values[0:400], '0.6', label='SAM original signal')
ax.plot(sam_gauss[0:400],'g',label=r'Gaussian filter, $\sigma$ = {}'.format(sigma),lw=1.)
plt.plot(sam_filt_butter[0:400], 'r', label='scipy.signal\nbutterworth filter')
ax.plot(sam_filtered[0:400],'b',label='Matlab\nbutterworth filter',lw=1.5)
ax.legend(loc=0, frameon=False)
ax.grid('on')
ax.set_title('{} - {} days'.format(high, low))
f.savefig('../../figures/data_processing/comparison_filtered_20_120_days_scipy_signal.png', dpi=200, bbox_inches='tight') 
```
